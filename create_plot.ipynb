```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

plt.rc('figure', figsize=(16.0, 12.0))
sns.set_context("talk")
sns.set_style("whitegrid")
```

Source: https://onepiece.fandom.com/wiki/SBS_Volume_73

Inspiration: https://www.reddit.com/r/OnePiece/comments/iavv1j/will_zoro_say_sanji_from_one_piece_party_3/g1scwe0?utm_source=share&utm_medium=web2x

```python
df = pd.read_csv("zoro_sanji - Sheet1.csv")
df = df[df.columns[:-2]]

df.Category = df.Category.map(lambda x: [y.strip() for y in x.split(",")])
df = df.explode("Category")
```

```python
pdf = df.groupby("English term")['Times'].sum().reset_index().sort_values("Times")
sns.barplot(y="English term", x="Times", data=pdf)
```

```python
pal = sns.color_palette(["#7fad86", "#faf679"])
```

```python
pdf = df.groupby(["Category", "Strawhat"])['Times'].sum().reset_index()
sns.barplot(y="Category", x="Times", hue="Strawhat", hue_order=["Zoro", "Sanji"], data=pdf, palette=pal)
sns.despine(bottom=True)
plt.ylabel("Insult type");
plt.xlabel("Frequency");
plt.savefig("test.svg")
```
